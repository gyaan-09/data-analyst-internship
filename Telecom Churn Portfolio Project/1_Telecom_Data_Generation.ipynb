```
import pandas as pd
import numpy as np

# 1. Initialization
np.random.seed(42)
num_rows = 10000

# 2. Base Features
customer_ids = [f"TEL-{10000 + i}" for i in range(num_rows)]
contract_types = ['Month-to-Month', '1-Year', '2-Year']
network_status = ['4G', '5G']

data = []

# 3. Data Generation with Embedded Churn Logic
for i in range(num_rows):
    # Inject ~4% NULLs for SQL Data Cleaning
    contract = np.random.choice(contract_types, p=[0.6, 0.25, 0.15]) if np.random.rand() > 0.04 else None
    network = np.random.choice(network_status, p=[0.7, 0.3])

    tenure_months = np.random.randint(1, 72)
    monthly_bill = np.random.uniform(20.0, 120.0)

    # Behavioral Metrics
    call_duration_mins = int(np.random.normal(loc=500, scale=150))
    recharge_frequency_days = int(np.random.normal(loc=28, scale=10))
    network_drops = int(np.random.exponential(scale=2))
    complaints = np.random.choice([0, 1, 2, 3, 4], p=[0.6, 0.2, 0.1, 0.07, 0.03])

    # Mathematical Logic: What causes churn?
    churn_prob = 0.05 # Baseline

    if contract == 'Month-to-Month': churn_prob += 0.25
    if network_drops > 4: churn_prob += 0.30
    if complaints >= 2: churn_prob += 0.20
    if network == '5G': churn_prob -= 0.15 # 5G retains customers
    if tenure_months > 24: churn_prob -= 0.10

    churn_prob = max(0.01, min(0.95, churn_prob))
    churn = 1 if np.random.rand() < churn_prob else 0

    # Force 'Dormant' behavior (No calls, high days since recharge)
    if tenure_months > 6 and np.random.rand() < 0.05:
        call_duration_mins = 0
        recharge_frequency_days = np.random.randint(60, 120)
        churn = 1

    call_duration_mins = max(0, call_duration_mins)
    recharge_frequency_days = max(1, recharge_frequency_days)

    data.append([
        customer_ids[i], tenure_months, contract, network, round(monthly_bill, 2),
        call_duration_mins, recharge_frequency_days, network_drops, complaints, churn
    ])

# 4. Export Raw Data
columns = ['customer_id', 'tenure_months', 'contract_type', 'network_type', 'monthly_bill',
           'total_call_duration', 'days_since_recharge', 'network_drops', 'complaints', 'churn_label']
df = pd.DataFrame(data, columns=columns)
df.to_csv('1_telecom_raw_data.csv', index=False)
print("SUCCESS: 1_telecom_raw_data.csv generated.")
```
